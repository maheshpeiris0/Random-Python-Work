```python
def diagram_stock(file_path):
    df=pd.read_json(file_path, lines=True)
    df['datetime_est']=pd.to_datetime(df['start_timestamp'],unit='ms',utc=True).dt.tz_convert('US/Eastern')
    fig, ax = plt.subplots()
    ax.plot(df['datetime_est'], df['close'])
    ax.set(xlabel='Date', ylabel='Close Price', title='Stock Price')
    ax.grid()
    plt.show()
```

```python
diagram_stock('/workspaces/Random-Python-Work/2024/March/RandomValuesGenerations/XTLB.json')
```
